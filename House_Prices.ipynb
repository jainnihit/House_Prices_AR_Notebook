```python
df_train = pd.read_csv("D:\Machine Learning\Kaggle\House _Prices_Advanced_Regression\\train.csv")
df_test = pd.read_csv("D:\Machine Learning\Kaggle\House _Prices_Advanced_Regression\\test.csv")
```

```python
df_trainWithoutOutliers = df_train[np.abs(df_train["SalePrice"]-df_train["SalePrice"].mean())<=(3*df_train["SalePrice"].std())]
```

```python
df_trainWithoutOutliers.shape
```

```python
df_trainWithoutOutliers.columns[df_trainWithoutOutliers.isnull().any()]
```

```python
df_test.columns[df_test.isnull().any()]
```

```python
del df_trainWithoutOutliers['PoolQC']
del df_test['PoolQC']
```

```python
del df_trainWithoutOutliers["MiscFeature"]
del df_test["MiscFeature"]
```

```python
df_trainWithoutOutliers['Alley']=df_trainWithoutOutliers['Alley'].fillna('NA')
df_test['Alley']=df_test['Alley'].fillna('NA')
```

```python
#Group by neighborhood and fill in missing value by the median LotFrontage of all the neighborhood
df_trainWithoutOutliers['LotFrontage'] = df_trainWithoutOutliers.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
df_test['LotFrontage'] = df_test.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
```

```python
#GarageType, GarageFinish, GarageQual and GarageCond these are replacing with None
for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond','Fence']:
    df_trainWithoutOutliers[col] = df_trainWithoutOutliers[col].fillna('NA')
    df_test[col] = df_test[col].fillna('NA')
    
#GarageYrBlt, GarageArea and GarageCars these are replacing with zero
for col in ['GarageYrBlt', 'GarageArea', 'GarageCars']:
    del df_trainWithoutOutliers[col]
    del df_test[col]
```

```python
df_trainWithoutOutliers['FireplaceQu'] = df_trainWithoutOutliers['FireplaceQu'].fillna(df_trainWithoutOutliers['FireplaceQu']).mode()[0]
df_test['FireplaceQu'] = df_test['FireplaceQu'].fillna(df_test['FireplaceQu']).mode()[0]
```

```python
#BsmtFinType2, BsmtExposure, BsmtFinType1, BsmtCond, BsmtQual these are replacing with None
for col in ('BsmtFinType2', 'BsmtExposure', 'BsmtFinType1', 'BsmtCond', 'BsmtQual'):
    df_trainWithoutOutliers[col] = df_trainWithoutOutliers[col].fillna('None'+col)    
    df_test[col] = df_test[col].fillna('None'+col)
```

```python
#MasVnrArea : replace with zero
df_trainWithoutOutliers['MasVnrArea'] = df_trainWithoutOutliers['MasVnrArea'].fillna(int(0))  
df_test['MasVnrArea'] = df_test['MasVnrArea'].fillna(int(0)) 
```

```python
df_trainWithoutOutliers['MasVnrType'] = df_trainWithoutOutliers['MasVnrType'].fillna('None')
df_test['MasVnrType'] = df_test['MasVnrType'].fillna('None')
```

```python
#There is put mode value 
df_trainWithoutOutliers['Electrical'] = df_trainWithoutOutliers['Electrical'].fillna(df_trainWithoutOutliers['Electrical']).mode()[0]
df_test['Electrical'] = df_test['Electrical'].fillna(df_test['Electrical']).mode()[0]
#There is no need of Utilities
df_trainWithoutOutliers = df_trainWithoutOutliers.drop(['Utilities'], axis=1)
df_test = df_test.drop(['Utilities'], axis=1)
```

```python
df_test['Exterior1st'] = df_test.groupby("Neighborhood")["Exterior1st"].transform(lambda x: x.fillna(x.mode()))

df_test['Exterior2nd'] = df_test.groupby("Neighborhood")["Exterior2nd"].transform(lambda x: x.fillna(x.mode()))

df_test['MSZoning'] = df_test.groupby("Neighborhood")["MSZoning"].transform(lambda x: x.fillna(x.mode()))
```

```python
df_test['SaleType'] = df_test['SaleType'].fillna('Oth')
```

```python
df_test['KitchenQual'] = df_test['KitchenQual'].fillna(df_test['KitchenQual']).mode()[0]
```

```python
df_test['BsmtHalfBath'] = df_test['BsmtHalfBath'].fillna(df_test['BsmtHalfBath']).mode()[0]
```

```python
df_test['BsmtFullBath'] = df_test['BsmtFullBath'].fillna(df_test['BsmtFullBath']).mode()[0]
```

```python
df_test['BsmtFinSF1'] = df_test["BsmtFinSF1"].transform(lambda x: x.fillna(x.median()))
```

```python
df_test['BsmtFinSF2'] = df_test['BsmtFinSF2'].transform(lambda x: x.fillna(x.median()))
```

```python
df_test['BsmtUnfSF'] = df_test["BsmtUnfSF"].transform(lambda x: x.fillna(x.median()))
```

```python
df_test['TotalBsmtSF'] = df_test["BsmtUnfSF"]+df_test['BsmtFinSF2']+df_test['BsmtFinSF1'] 
```

```python
df_test['Functional'] = df_test['Functional'].fillna('Typ')
```

```python
df_trainWithoutOutliers = df_trainWithoutOutliers.drop(['Id'], axis = 1)
ID = df_test['Id']
df_test = df_test.drop(['Id'], axis = 1)
```

```python
df_trainWithoutOutliers.info()
```

```python
cols = df_trainWithoutOutliers.select_dtypes(object)
from sklearn.preprocessing import LabelEncoder
for c in cols:
    LE = LabelEncoder() 
    LE.fit(list(df_trainWithoutOutliers[c].values)) 
    df_trainWithoutOutliers[c] = LE.transform(list(df_trainWithoutOutliers[c].values))
for c in cols:
    LE = LabelEncoder() 
    LE.fit(list(df_test[c].values)) 
    df_test[c] = LE.transform(list(df_test[c].values))
```

```python
LE = LabelEncoder()
LE.fit(list(df_trainWithoutOutliers['MSSubClass'].values)) 
df_trainWithoutOutliers['MSSubClass'] = LE.transform(list(df_trainWithoutOutliers['MSSubClass'].values))
LE.fit(list(df_test['MSSubClass'].values)) 
df_test['MSSubClass'] = LE.transform(list(df_test['MSSubClass'].values))
```

```python
Y_train = df_trainWithoutOutliers['SalePrice']
del df_trainWithoutOutliers['SalePrice']

#Take their values in X and y
X = df_trainWithoutOutliers.values
Y = Y_train.values

# Split data into train and test formate
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
from sklearn.preprocessing import StandardScaler
X_train = StandardScaler().fit_transform(X_train)
X_test = StandardScaler().fit_transform(X_test)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
model.fit(X_train, y_train)
y_pred_XGBR = model.predict(df_test.values) 
```

```python
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import RobustScaler
ENet = make_pipeline(RobustScaler(), ElasticNet(alpha=0.0005, l1_ratio=.9, random_state=3))
ENet.fit(X_train, y_train)
y_pred_ENET = ENet.predict(df_test.values)
```

```python
submission = pd.DataFrame({
        "Id": ID,
        "SalePrice": y_pred_ENET
    })
```
